```python
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# lodaing data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()

#data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
```

```python
df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()
```

```python
df_DA
```

```python
df_DA['job_posted_month_no'] = df_DA['job_posted_date'].dt.month

df_DA
```

```python
df_DA_exploded = df_DA.explode('job_skills')

df_DA_exploded
```

```python
df_DA_pivot = df_DA_exploded.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)

df_DA_pivot
```

```python
df_DA_pivot.loc['Total'] = df_DA_pivot.sum()

df_DA_pivot
```

```python
df_DA_pivot.loc['Total'].sort_values(ascending=False)
```

```python
df_DA_pivot.loc['Total'].sort_values(ascending=False).index
```

```python
df_DA_pivot = df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending=False).index]

df_DA_pivot
```

```python
df_DA_pivot = df_DA_pivot.drop('Total')

df_DA_pivot
```

```python
df_DA_pivot.iloc[:, :5]
```

```python
df_DA_pivot.iloc[:, :5].plot(kind='line')
plt.title('Top 5 Trending Skills for Data Analysts Over Months')
plt.xlabel('Month Number')
plt.ylabel('Number of Job Postings')
plt.show
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Reset index
df_DA_pivot = df_DA_pivot.reset_index()

# Convert month number → month name
df_DA_pivot['job_posted_month'] = df_DA_pivot['job_posted_month_no'].apply(
    lambda x: pd.to_datetime(x, format='%m').strftime('%B')
)

# Set month name as index
df_DA_pivot = df_DA_pivot.set_index('job_posted_month')

# Drop the numeric month column
df_DA_pivot = df_DA_pivot.drop(columns='job_posted_month_no')

# Plot top 5 skills per month
df_DA_pivot.iloc[:, :5].plot(kind="line")

plt.title('Top 5 Skills for Data Analysts per Month')
plt.ylabel('Count')
plt.xlabel('Month')
plt.show()
```
