### Exploratory Data Analysis for all Data Roles

#### Roles to Explore

```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt


# lodaing data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()

#data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
```

```python
df_plot = df['job_title_short'].value_counts().to_frame()

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_title_short', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Number of Jobs per Job Title')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

#### Countries to Explore

```python
df_plot = df['job_country'].value_counts().to_frame().head(20)

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_country', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Number of Jobs per Country')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

#### Companies to Explore

```python
df_plot = df['company_name'].value_counts().to_frame()[1:].head(20)

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='company_name', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Number of Jobs per Company')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

#### Job Opportunities

```python
dict_column = {
    'job_work_from_home': 'Work from Home Offered',
    'job_no_degree_mention': 'Degree Requirement',
    'job_health_insurance': 'Health Insurance Offered'
}

fig, ax = plt.subplots(1, 3, figsize=(11, 3.5))

for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(df[column].value_counts(), labels=['False', 'True'], autopct='%1.1f%%', startangle=90)
    ax[i].set_title(title)

plt.show()
```

#### Exploratory Data Analysis for Data Scientist and Data Analysts in the Uk

```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt


# lodaing data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()

#data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
```

#### Filter for Uk Data Scientist roles

```python
df_DS_UK = df[(df['job_country'] == 'United Kingdom') & (df['job_title_short'] == 'Data Scientist')]

df_DS_UK
```

#### Filter for Uk Data Analyst roles

```python
df_DA_UK = df[(df['job_country'] == 'United Kingdom') & (df['job_title_short'] == 'Data Analyst')]

df_DA_UK
```

#### Locations to Explore:

```python
df_plot = df_DS_UK['job_location'].value_counts().head(10).to_frame()

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_location', hue='count', palette="ch:s=-.2,r=.6,l=.75", legend=False)
sns.despine()
plt.title('Counts of Job Locations for Data Scientist in the UK')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

```python
df_plot = df_DA_UK['job_location'].value_counts().head(10).to_frame()

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_location', hue='count', palette="viridis_r", legend=False)
sns.despine()
plt.title('Counts of Job Locations for Data Analyst in the UK')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

#### Job Opportunities

```python
dict_column = {
    'job_work_from_home': 'Work from Home Offered',
    'job_no_degree_mention': 'Degree Requirement',
    'job_health_insurance': 'Health Insurance Offered'
}

# make subplots
fig, ax = plt.subplots(1, 3, figsize=(14, 5))

palette = sns.color_palette("Set2", 2)   

for i, (column, title) in enumerate(dict_column.items()):
    values = df_DS_UK[column].value_counts()
    ax[i].pie(
        values,
       labels=['False', 'True'],
        autopct='%1.1f%%',
        startangle=90,
        colors=palette
    )
    ax[i].set_title(title, fontsize=12)

plt.suptitle("Benefit Analysis of Data Science Jobs in the UK", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
dict_column = {
    'job_work_from_home': 'Work from Home Offered',
    'job_no_degree_mention': 'Degree Requirement',
    'job_health_insurance': 'Health Insurance Offered'
}

fig, ax = plt.subplots(1, 3, figsize=(14, 5))

colors = sns.color_palette(["#4C72B0", "#8E44AD"])

for i, (column, title) in enumerate(dict_column.items()):
    values = df_DA_UK[column].value_counts()
    ax[i].pie(
        values,
        labels=['False', 'True'],
        autopct='%1.1f%%',
        startangle=90,
        colors=colors
    )
    ax[i].set_title(title, fontsize=12)

plt.suptitle("Benefit Analysis of Data Analyst Jobs in the UK", fontsize=16)
plt.tight_layout()
plt.show()
```
